# Consolidated benchmark summary

This notebook runs `scripts/consolidate_benchmarks.py` against `results/` and prints a human-readable rollup (compress/decompress, cosine quality, rocprof latency JSON, MFMA rotation bench).

**Prereqs:** run benchmarks first so JSON files exist under `results/`.

```python
from pathlib import Path
import subprocess, sys

root = Path.cwd().resolve()
if not (root / "results").exists():
    root = root.parent  # if launched from notebooks/
script = root / "scripts" / "consolidate_benchmarks.py"
assert script.exists(), script

cmd = [sys.executable, str(script), "--verbose", "--write-json", str(root / "results" / "consolidated_benchmarks.json")]
print(" ".join(cmd))
subprocess.run(cmd, cwd=str(root), check=True)
```

```python
from pathlib import Path
import json

p = Path("../results/consolidated_benchmarks.json").resolve()
if p.exists():
    data = json.loads(p.read_text())
    print("keys:", list(data.keys()))
    print("files recorded:", len(data.get("files", {})))
else:
    print("Run previous cell first —", p)
```
